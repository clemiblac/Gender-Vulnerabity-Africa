```python
#Cause of death
gender_cause_of_death=gender_related_topics[gender_related_topics['name'].str.contains('Cause of death')]
get_indicator_id=gender_cause_of_death.reset_index()
series=get_indicator_id['id']
```

```python
id_list=series.tolist()
id_list
```

```python
gender_cause_of_death['name'].unique()
```

```python
#wb.search_regions('Africa')
```

```python
final=[]
for i in id_list:
    #print(i)
    list_name='series_'+i
    ls=list_name.replace('.', '_')
    print(ls)
    data={ls:wb.get_series(i,date='2016', id_or_value='value', simplify_index=True)}
    final.append(data)
    
    
```

```python
final
```

```python
series_SH_DTH_COMM_1534_FE_ZS = final[0]
series_SH_DTH_COMM_1534_MA_ZS = final[1]
series_SH_DTH_COMM_3559_FE_ZS = final[2]
series_SH_DTH_COMM_3559_MA_ZS = final[3]
series_SH_DTH_INJR_1534_FE_ZS = final[4]
series_SH_DTH_INJR_1534_MA_ZS = final[5]
series_SH_DTH_INJR_3559_FE_ZS = final[6]
series_SH_DTH_INJR_3559_MA_ZS = final[7]
series_SH_DTH_NCOM_1534_FE_ZS = final[8]
series_SH_DTH_NCOM_1534_MA_ZS = final[9]
series_SH_DTH_NCOM_3559_FE_ZS = final[10]
series_SH_DTH_NCOM_3559_MA_ZS = final[11]
```

```python
SH_DTH_COMM_1534_FE_ZS_df=pd.DataFrame(series_SH_DTH_COMM_1534_FE_ZS)
SH_DTH_COMM_1534_MA_ZS_df=pd.DataFrame(series_SH_DTH_COMM_1534_MA_ZS)
SH_DTH_COMM_3559_FE_ZS_df=pd.DataFrame(series_SH_DTH_COMM_3559_FE_ZS)
SH_DTH_COMM_3559_MA_ZS_df=pd.DataFrame(series_SH_DTH_COMM_3559_MA_ZS)

SH_DTH_INJR_1534_FE_ZS_df=pd.DataFrame(series_SH_DTH_INJR_1534_FE_ZS)
SH_DTH_INJR_1534_MA_ZS_df=pd.DataFrame(series_SH_DTH_INJR_1534_MA_ZS)
SH_DTH_INJR_3559_FE_ZS_df=pd.DataFrame(series_SH_DTH_INJR_3559_FE_ZS)
SH_DTH_INJR_3559_MA_ZS_df=pd.DataFrame(series_SH_DTH_INJR_3559_MA_ZS)

SH_DTH_NCOM_1534_FE_ZS_df=pd.DataFrame(series_SH_DTH_NCOM_1534_FE_ZS)
SH_DTH_NCOM_1534_MA_ZS_df=pd.DataFrame(series_SH_DTH_NCOM_1534_MA_ZS)
SH_DTH_NCOM_3559_FE_ZS_df=pd.DataFrame(series_SH_DTH_NCOM_3559_FE_ZS)
SH_DTH_NCOM_3559_MA_ZS_df=pd.DataFrame(series_SH_DTH_NCOM_3559_MA_ZS)
```

```python
SH_DTH_COMM_1534_FE_ZS_df.tail(5)
```

```python
### Merging all the tables into 1
#SH_DTH_COMM_1534_FE_ZS_df, SH_DTH_COMM_1534_MA_ZS_df, SH_DTH_COMM_3559_FE_ZS_df, SH_DTH_COMM_3559_MA_ZS_df,
#SH_DTH_INJR_1534_FE_ZS_df, SH_DTH_INJR_1534_MA_ZS_df, SH_DTH_INJR_3559_FE_ZS_df, SH_DTH_INJR_3559_MA_ZS_df,
#SH_DTH_NCOM_1534_FE_ZS_df, SH_DTH_NCOM_1534_MA_ZS_df, SH_DTH_NCOM_3559_FE_ZS_df, SH_DTH_NCOM_3559_MA_ZS_df
```

```python
comm_0=pd.merge(SH_DTH_COMM_1534_FE_ZS_df, SH_DTH_COMM_1534_MA_ZS_df, on="Country")
comm_1=pd.merge(SH_DTH_COMM_3559_FE_ZS_df, SH_DTH_COMM_3559_MA_ZS_df, on="Country")
communicable=pd.merge(comm_0,comm_1, on="Country")
communicable_rn=communicable.rename(columns={"series_SH_DTH_COMM_1534_FE_ZS": "COMM_1534_FE_per",
                                             "series_SH_DTH_COMM_1534_MA_ZS": "COMM_1534_MA_per",
                                             "series_SH_DTH_COMM_3559_FE_ZS": "COMM_3559_FE_per",
                                             "series_SH_DTH_COMM_3559_MA_ZS": "COMM_3559_MA_per"})
communicable_rn.head(4)
```

```python
non_comm_0=pd.merge(SH_DTH_NCOM_1534_FE_ZS_df, SH_DTH_NCOM_1534_MA_ZS_df, on="Country")
non_comm_1=pd.merge(SH_DTH_NCOM_3559_FE_ZS_df, SH_DTH_NCOM_3559_MA_ZS_df, on="Country")
non_communicable=pd.merge(non_comm_0,non_comm_1, on="Country")
non_communicable_rn=non_communicable.rename(columns={"series_SH_DTH_NCOM_1534_FE_ZS":"NCOM_1534_FE_per",
                                                     "series_SH_DTH_NCOM_1534_MA_ZS":"NCOM_1534_MA_per",
                                                     "series_SH_DTH_NCOM_3559_FE_ZS":"NCOM_3559_FE_per",
                                                     "series_SH_DTH_NCOM_3559_MA_ZS":"NCOM_3559_MA_per"})

non_communicable_rn.head(4)
```

```python
inj_0=pd.merge(SH_DTH_INJR_1534_FE_ZS_df, SH_DTH_INJR_1534_MA_ZS_df,on="Country")
inj_1=pd.merge(SH_DTH_INJR_3559_FE_ZS_df, SH_DTH_INJR_3559_MA_ZS_df, on="Country")
injury=pd.merge(inj_0,inj_1, on="Country")
injury_rn=injury.rename(columns={"series_SH_DTH_INJR_1534_FE_ZS":"INJR_1534_FE_per",
                                 "series_SH_DTH_INJR_1534_MA_ZS":"INJR_1534_MA_per",
                                 "series_SH_DTH_INJR_3559_FE_ZS":"NJR_3559_FE_per",
                                 "series_SH_DTH_INJR_3559_MA_ZS":"INJR_3559_MA_per"})

injury_rn.head(4)
```

```python
intermediate_1=pd.merge(communicable_rn,non_communicable_rn,on='Country')
intermediate_2=pd.merge(intermediate_1,injury_rn,on='Country')
data=pd.merge(Sub_Saharan_Africa,intermediate_2,on="Country")
```

```python
data.head(5)
```

```python
json_wb_deaths = data.to_json(orient = 'records')
parsed = json.loads(json_wb_deaths)
```

```python
with open('cause_of_death_2016.json', 'w') as f:
    json.dump(parsed, f, indent=4)
```
